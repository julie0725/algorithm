```python
import matplotlib.pyplot as plt
import csv
```

```python
# CSV 파일 읽기
n_values = []
m_values = []
brute_force_times = []
kmp_times = []

with open('C:\\Users\\juyeo\\Desktop\\ACCGTAT\\time_results.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # 헤더 건너뛰기
    for row in reader:
        n_values.append(int(row[0]))  # n 값
        m_values.append(int(row[1]))  # m 값
        brute_force_times.append(float(row[2]))  # Brute-Force 실행 시간
        kmp_times.append(float(row[3]))  # KMP 실행 시간
```

```python
# 그래프 그리기
plt.figure(figsize=(10,6))
plt.plot(n_values, brute_force_times, marker='o', label="Brute-Force", color='b')
plt.plot(n_values, kmp_times, marker='o', label="KMP", color='r')
plt.xlabel('File Size (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Brute-Force vs KMP Execution Time')
plt.legend()  # 범례 추가
plt.grid(True)
plt.show()
```
